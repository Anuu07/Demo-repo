```python
import cv2
import matplotlib.pyplot as plt

```

```python
# Start the webcam
cap = cv2.VideoCapture(0)

# Capture one frame
ret, frame = cap.read()

# Stop the webcam
cap.release()

# Show the captured image
if ret:
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame_rgb)
    plt.axis('off')
    plt.title("Captured Image")
    plt.show()
else:
    print("Failed to access webcam.")
```
